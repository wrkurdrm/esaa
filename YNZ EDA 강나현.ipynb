```python
ax = plt.subplots() 
ax = sn.countplot('cus_age', data=cus_info) 
ax.set_title('연령대 분포') 
ax.set_xlabel('연령대 (00: 19세 이하, 20: 20~24,~ 70: 70세 이상)') 
ax.set_ylabel('Frequency') 
```

Y&Z 세대 투자자 프로파일링이라서 X,Y,Z 세대가 각각 무슨 뜻인지 알려주는 그림 첨부

데이콘 데이터 명세에는 20년 6월 기준 자료라고 되어 있으므로 나이는 2020년 기준으로 계산. 만 나이로
즉 Z세대는 00,20, Y세대는 25, 30, 35, X세대는 40, 45, 50 범주이다. 추후 분석에서 00과 20을 한 범주(Z)로, 25~35를 한 범주(Y)로 다시 묶어 새로운 변수를 만든다면 주제에 맞는 분석에 유용할 것이라 생각.

## [HTS와 MTS에 대한 설명 기사](https://www.cidermics.com/contents/detail/1874)

동학 개미는 2020년 초 코로나19로 인해 외국인 투자자들이 주식 매도를 벌이자, 이에 대응하여 주식을 적극적으로 매수한 개인투자자를 의미합니다. ... 국내 주식을 사는 동학개미를 넘어 애플, 테슬라 등 미국 주식에 적극적으로 투자하는 개미를 서학개미라고 합니다.

MZ세대란, 1980년부터 2004년생까지를 일컫는 밀레니얼 세대와 1995년부터 2004년 출생자를 뜻하는 Z세대를 합쳐 일컫는 말입니다. 통계청에 따르면 이 MZ세대는 2019년 기준 약 1,700만 명으로 국내 인구의 약 34%를 차지한다고 합니다. (MZ세대=Y&Z세대라고 보면 될듯)
